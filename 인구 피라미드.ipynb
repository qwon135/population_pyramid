# 1. 인구 피라미드

### 남자 데이터 불러오기
필요없는 row를 자르고 usecols로 필요한 column만 뽑아오기

```python
import pandas as pd
file_name='202108_202108_연령별인구현황_월간.xlsx'
df_m=pd.read_excel(file_name,skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m.head(3)
```

### 숫자에 ' , ' 없애기

```python
df_m.iloc[0]=df_m.iloc[0].str.replace(',','').astype(int) # 콤마를 빈 문자열로 교체 후 astype 을 사용해 정수처리
df_m
```

## 여자 데이터 불러오기

```python
import pandas as pd
df_w=pd.read_excel(file_name,skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns # 남자데이터의 column 값과 여자 데이터의 colum이 일치해서 뒤에 .1이 붙게된다.
```

```python
df_w.columns=df_m.columns # 남,여의 column 데이터 통일
```

```python
df_w.iloc[0]=df_w.iloc[0].str.replace(',','').astype(int) # 똑같이 콤마를 없애고 정수형으로 바꿔준다.
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=12
matplotlib.rcParams['axes.unicode_minus']=False 
```

### 막대 그래프

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns,-df_m.iloc[0]//1000) # 단 위 : 천 명 / 남자 데이터는 음수
plt.barh(df_w.columns,df_w.iloc[0]//1000) # 여자데이터는 양수
plt.title('2021년 대한민국 인구 피라미드')
plt.savefig('2021_인구피라미드.png',dpi=100)
plt.show()
```
